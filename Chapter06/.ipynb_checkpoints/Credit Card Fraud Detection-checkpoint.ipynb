```python
import pandas as pd
df = pd.read_csv("creditcard.csv", index_col=None)
```

```python
df.head()
```

```python
from collections import Counter
Counter(df["Class"].values)
```

```python
l = len(df.index)
train_df = df.head(int(l*0.8))
test_df = df.tail(int(l*0.2))
```

```python
print(Counter(train_df["Class"].values))
print(Counter(test_df["Class"].values))
```

```python
y_train = train_df.pop("Class").values
y_test = test_df.pop("Class").values
```

```python
X_train = train_df.values
X_test = test_df.values
```

```python
X_train_normal = X_train[y_train==0]
print(X_train_normal.shape)
X_train_threat = X_train[y_train==1]
print(X_train_threat.shape)
X_test_normal = X_test[y_test==0]
print(X_test_normal.shape)
X_test_threat = X_test[y_test==1]
print(X_test_threat.shape)
```

```python
from sklearn.ensemble import IsolationForest
contaminationParameter = sum(y_train==1)/sum(y_train==0)
IF = IsolationForest(n_estimators=100, max_samples=256,contamination=contaminationParameter)
```

```python
IF.fit(X_train)
```

```python
import matplotlib.pyplot as plt
normalScores = IF.decision_function(X_train_normal)
fig = plt.figure(figsize=(8, 4), dpi=600, facecolor='w', edgecolor='k')

normal = plt.hist(normalScores, 50,density=True)

plt.xlabel('Anomaly score')
plt.ylabel('Percentage')
plt.title("Distribution of anomaly score for benign transactions")
```

```python
anomalyScores = IF.decision_function(X_train_threat)
fig = plt.figure(figsize=(8, 4), dpi=600, facecolor='w', edgecolor='k')

normal = plt.hist(anomalyScores, 50,density=True)

plt.xlabel('Anomaly score')
plt.ylabel('Percentage')
plt.title("Distribution of anomaly score for fraud transactions")
```

```python
cutoff = -0.05
```

```python
s = IF.decision_function(X_train)
print(Counter(y_train[cutoff>s]))
```

```python
s = IF.decision_function(X_test)
print(Counter(y_test[cutoff>s]))
```
